```python
#random forest classifier meeans multiple desicion trees - collection of multiple tress - it works with
#multiple desicion tree with random data together it will return onr final ans with majority votes
#10 students checking pass or fail in that 7 friends are saying pass 3 are saying fail majority is pass so the result is pass
```

```python
from sklearn.ensemble import RandomForestClassifier #ensemble - multiple trees
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = { 'Wheather' : ['sunny','sunny','overcast','rainy','rainy'],'play' : ['no','no','yes','yes','yes']}
df = pd.DataFrame(data)
```

```python
df
```

```python
df['play'].map({'yes' : 1, 'no' : 0})
#with this map function directly it will show yes means 1, no means 0 with in the dictionary format
```

```python
df['Wheather']=LabelEncoder().fit_transform(df['Wheather'])
```

```python
x = df[['Wheather']]
y = df['play']
```

```python
rfc = RandomForestClassifier(n_estimators=10,random_state=0)
```

```python
rfc
```

```python
rfc.fit(x,y)
```

```python
rfc.predict([[0]])
```

```python
print(rfc.predict(pd.DataFrame([[1]],
            columns = ['Wheather'])))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
s,e = map(int,input("enter the tree range ex : (0 2)").split())

# Ensure s and e are within valid bounds (0 to rfc.n_estimators - 1)
s = max(0, s)
e = min(rfc.n_estimators - 1, e)

for i in range(s, e + 1):
  plt.figure(figsize = (4,5))
  tree.plot_tree(
      rfc.estimators_[i],
      feature_names=['Wheather'], # Corrected feature name casing
      class_names=['no','yes'], # Corrected class names order (0: no, 1: yes)
      rounded = True,
      filled=True
  )
  plt.title(f"tree no {i}")
  plt.show() # Display each plot
```
